```python
from scrape import Scraper
from clean_data import DataCleaner
import pandas as pd
```

```python
scrape = Scraper()
```

```python
datasets = pd.DataFrame()
for key, value in scrape.BANK_APPS.items():
    df = scrape.scrape_bank_reviews(value, key)
    datasets = pd.concat([datasets, df], ignore_index=True)    
```

```python
datasets = datasets.drop_duplicates(subset=None, keep='first', inplace=False)
datasets.duplicated().sum()
```

```python
datasets.to_csv("../data/reviews_dataset.csv", index=False)
```
